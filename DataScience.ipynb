```python
import pandas as pd

# Carregar os dados
df_suicidio = pd.read_csv("datasus_suicidio_2014_2018.csv", encoding='ISO-8859-1')

# Exibir as primeiras linhas do DataFrame
print(df_suicidio.head())
```

```python
import matplotlib.pyplot as plt

# Agrupar por ano e contar o número de suicídios
suicidios_por_ano = df_suicidio.groupby('ano').size()

# Plotar o gráfico de linha
plt.plot(suicidios_por_ano.index, suicidios_por_ano.values, marker='o')
plt.title('Número de suicídios ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('Número de suicídios')
plt.grid(True)
plt.show()
```
