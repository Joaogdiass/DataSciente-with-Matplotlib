```python
import pandas as pd

# Carregar os dados
df_suicidio = pd.read_csv("datasus_suicidio_2014_2018.csv", encoding='ISO-8859-1')

```

1- Como o número de suicídios mudou ao longo dos anos?

```python
import matplotlib.pyplot as plt

#Agrupar por ano e contar o número de suicídios
suicidios_por_ano = df_suicidio.groupby('ano').size()

#Plotar o gráfico de linha
plt.plot(suicidios_por_ano.index, suicidios_por_ano.values, marker='o')
plt.title('Número de suicídios ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('Número de suicídios')
plt.grid(True)
plt.show()
```

2- Qual é a distribuição dos suicídios por estado?

```python
#Contar o número de suicídios por estado
suicidios_por_estado = df_suicidio['estado'].value_counts()

#Plotar o gráfico de barras horizontal
plt.barh(suicidios_por_estado.index, suicidios_por_estado.values)
plt.title('Distribuição de suicídios por estado')
plt.xlabel('Número de suicídios')
plt.gca().invert_yaxis()
plt.ylabel('Estado')
plt.show()
```

3- Existe uma diferença significativa entre os sexos em relação ao número de suicídios?


```python
#Contar o número de suicídios por sexo
suicidios_por_sexo = df_suicidio['SEXO'].value_counts()

#Plotar o gráfico de pizza
plt.pie(suicidios_por_sexo, labels=suicidios_por_sexo.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de suicídios por sexo')
plt.show()
```

4- Qual é a distribuição dos suicídios por faixa etária?

```python
# Contar o número de suicídios por faixa etária
suicidios_por_idade = df_suicidio['idade'].value_counts().sort_index()

# Plotar o histograma
plt.hist(df_suicidio['idade'], bins=range(0, 101, 5), edgecolor='black')
plt.title('Distribuição de suicídios por faixa etária')
plt.xlabel('Idade')
plt.ylabel('Número de suicídios')
plt.grid(True)
plt.show()
```

5- Há alguma correlação entre o estado civil e o número de suicídios?


```python
# Contar o número de suicídios por estado civil
suicidios_por_estadocivil = df_suicidio['ESTCIV'].value_counts()

# Plotar o gráfico de barras
plt.bar(suicidios_por_estadocivil.index, suicidios_por_estadocivil.values)
plt.title('Distribuição de suicídios por estado civil')
plt.xlabel('Estado Civil')
plt.ylabel('Número de suicídios')
plt.show()
```

6- Existe uma relação entre a escolaridade e o número de suicídios?

```python
# Contar o número de suicídios por nível de escolaridade
suicidios_por_escolaridade = df_suicidio['ESC'].value_counts()

# Plotar o gráfico de barras
plt.bar(suicidios_por_escolaridade.index, suicidios_por_escolaridade.values)
plt.title('Distribuição de suicídios por escolaridade')
plt.xlabel('Escolaridade')
plt.ylabel('Número de suicídios')
plt.show()
```

7- Como é a distribuição dos suicídios por ocupação? (O número 0 quer dizer que não se encaixa em nenhuma ocupação)




```python
# Contar o número de suicídios por ocupação
suicidios_por_ocupacao = df_suicidio['OCUP'].value_counts().head(15)

# Plotar o gráfico de barras horizontal
plt.barh(suicidios_por_ocupacao.index, suicidios_por_ocupacao.values)
plt.title('Distribuição dos suicídios por ocupação (Top 15)')
plt.xlabel('Número de suicídios')
plt.ylabel('Ocupação')
plt.gca().invert_yaxis()  # Inverter a ordem das barras para que a ocupação com mais suicídios fique no topo
plt.show()


```
